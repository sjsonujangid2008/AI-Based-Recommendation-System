```python
import pandas as pd

# -----------------------------
# Step 1: Create Sample Customer Data
# -----------------------------
data = {
    "CustomerID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Mobile","Laptop","Headphones","Mobile","Laptop","Watch","Mobile"],
    "Rating": [5,4,5,3,4,5,2,5]
}

df = pd.DataFrame(data)

print("Customer Purchase Data")
print(df)

# -----------------------------
# Step 2: Pattern Learning
# -----------------------------
print("\nMost Popular Products")
print(df["Product"].value_counts())

# -----------------------------
# Step 3: Recommendation Logic
# -----------------------------
popular_product = df["Product"].mode()[0]

print("\nRecommended Product")
print("Recommend:", popular_product)

# -----------------------------
# Step 4: Customer Recommendation
# -----------------------------
print("\nCustomer Recommendations")

for customer in df["CustomerID"]:
    print("Customer", customer, "->", popular_product)

# -----------------------------
# Step 5: Evaluation Strategy
# -----------------------------
print("\nEvaluation Strategy")
print("- Check customer satisfaction after recommendations.")
print("- Monitor repeat purchases.")
print("- Compare sales before and after recommendations.")

print("\nBusiness Impact")
print("A recommendation system helps increase sales and improve customer experience.")

print("\nCapstone Project 4 Completed Successfully!")
```
